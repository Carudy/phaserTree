```python
import numpy as np
from scipy.interpolate import CubicSpline, PchipInterpolator
import matplotlib.pyplot as plt

# Define the data points
x = np.array([0, 1, 2, 2 + 1e-6, 3])
y = np.array([1, 2, 3, 1, 0])

# Create the cubic spline interpolator
cs = PchipInterpolator(x, y)

# Generate points to plot the interpolating curve
xs = np.linspace(-1, 4, 1000)
ys = cs(xs)

# Plot the data points and the interpolating curve
plt.plot(x, y, 'o', label='Data')
plt.plot(xs, ys, label='Cubic Spline')
plt.legend()
plt.show()
```
